```python
import pandas as pd
```

```python
df = pd.read_csv('emissions.csv')
df.head()
```

```python
countries = df[['Country', 'Year', 'Emissions.Type.CO2', 'Emissions.Type.N2O', 'Emissions.Type.CH4']]
countries.to_csv('countries.csv', index=False)
```

```python
sectors_data = df[['Year', 'Emissions.Sector.Power Industry', 'Emissions.Sector.Buildings',
                   'Emissions.Sector.Transport', 'Emissions.Sector.Other Industry', 'Emissions.Sector.Other sectors']].set_index('Year')
```

```python
sectors_data = sectors_data.reset_index().melt(id_vars='Year', var_name='Sector', value_name='Emissions')
```

```python
sectors_data.head()
```

```python
sectors_data = sectors_data.groupby(['Sector', 'Year'], as_index=False).agg('sum')
```

```python
sectors_data.head()
```

```python
sectors_data.to_csv('sectors.csv', index=False)
```
